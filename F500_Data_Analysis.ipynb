```python
opened_webpage = requests.get("https://money.cnn.com/magazines/fortune/global500/2013/full_list/?iid=G500_sp_fulls")
print ("Webpage opened successfully...")

# Initialize a BeautifulSoup object to read and parse the webpage read
# This is like calling the __init__ function in BeautifulSoup
bs = BeautifulSoup(opened_webpage.content, "html.parser")
# above added in html parser because we want to parse the html this is important becasuse there are other kinds of parsers.
print ("Webpage loaded and parsed successfully...")
opened_webpage.content
```

```python
raw_data = []

results = bs.find_all('span', class_="listCol1")
results += bs.find_all('span', class_="listCol2")
results += bs.find_all('span', class_="listCol3")
results += bs.find_all('span', class_="profits-visible")
results += bs.find_all('span', class_="assets-visible")
print(results)
```

# Bivariate Analysis

Next I am going to analyze the relationship between rank and revenue to try and assess how the rankings of F500 companies is established.

```python
from scipy.stats import pearsonr

p = pearsonr(F500_Companies_df["Rank"],F500_Companies_df["Revenue"])
print (p[0])
```

```python
from scipy.stats import pearsonr

p = pearsonr(F500_Companies_df["Rank"],F500_Companies_df["Profits"])
print (p[0])
```

Next I am going to create a bar graph for the top 2 and bottom 2 companies on the list and compare their profits in the year 2013.

```python
top_bottom_two = [0,1,498,499]

company_column = F500_Companies_df["Company Name"]
profits_column = F500_Companies_df["Profits"]

selected_companies = []
selected_profits = []

for company_id in top_bottom_two:
    selected_companies.append(company_column[company_id])
    selected_profits.append(profits_column[company_id])
    

# Now plot our bar graph
plt.bar(selected_companies, selected_profits, color ='maroon', width = 0.4)

# Now add axis labels with units
plt.xlabel("Companies")
plt.ylabel("Profits in Billion $")

# We can add a title too
plt.title("Top 2 and Bottom 2 Fortune 500 Company Profits in 2013")
```

This graph highlights the difference in profits for the bottom two and top two fortune 500 companies as you can see it is a very large difference.
